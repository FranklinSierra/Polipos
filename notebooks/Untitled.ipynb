```python
import matplotlib.pyplot as plt

# Data to plot
labels = 'Sobreviven', 'Fallecen'
sizes = [39000,32000]
colors = ['gold', 'lightcoral']
#explode = ()  # explode 1st slice

# Plot
plt.pie(sizes, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
